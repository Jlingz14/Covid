```python
from math import sqrt
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
def  brownian ( x0 ,m,  n ,  dt ,  delta):
    x = np.zeros((m,n))
    x[0] = x0
    # Iterate to compute the steps of the Brownian motion.
    for j in range(m):
        for k in range(n):
            x[j,k] = x[j,k-1] + norm.rvs(scale=delta**2*dt)
    return x
```

```python
x = brownian(0,50,200,0.1,0.25)
for j in range(50):
    plt.plot(x[j,:])
```

```python
plt.plot(x[1,:],x[2,:])
plt.plot(x[1,0],x[2,0], 'go')
plt.plot(x[1,-1], x[2,-1], 'ro')
```
